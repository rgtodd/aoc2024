## Load Source Data

Load the map data into `DATA`.

## Parse Data



```python
KEYS = []
LOCKS = []

for row in range(0, len(DATA), 8):
    schematic = [0, 0, 0, 0, 0]
    for schematic_line in DATA[row + 1 : row + 6]:
        for col in range(5):
            if schematic_line[col] == "#":
                schematic[col] += 1
    if DATA[row] == "#####":
        LOCKS.append(tuple(schematic))
    else:
        KEYS.append(tuple(schematic))

for key in KEYS:
    print(f"key = {key}")
for lock in LOCKS:
    print(f"lock = {lock}")
```

## Analyze Locks

```python
LOCK_COUNT_SETS = {
    0: {0: set(), 1: set(), 2: set(), 3: set(), 4: set(), 5: set()},
    1: {0: set(), 1: set(), 2: set(), 3: set(), 4: set(), 5: set()},
    2: {0: set(), 1: set(), 2: set(), 3: set(), 4: set(), 5: set()},
    3: {0: set(), 1: set(), 2: set(), 3: set(), 4: set(), 5: set()},
    4: {0: set(), 1: set(), 2: set(), 3: set(), 4: set(), 5: set()},
}

for lock in LOCKS:
    for col in range(5):
        col_dict = LOCK_COUNT_SETS[col]
        
        for key_height in range(0, 5 - lock[col] + 1):
            col_dict[key_height].add(lock)
```

## Define find_locks

```python
def find_locks(key):
    locks = set(LOCKS)
    for idx in range(5):
        locks &= LOCK_COUNT_SETS[idx][key[idx]]
    return locks
```

## Find lock combinations

```python
count = 0
for key in KEYS:
    count += len(find_locks(key))

print(f"count = {count}")
```
