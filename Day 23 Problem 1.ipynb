## Load Source Data

Load the map data into `DATA`.

```python
f = open("data/day23.txt", "r")
DATA = list(map(str.strip, f.readlines()))
f.close()
```

## Construct and Render Graph



```python
import matplotlib.pyplot as plt
import igraph as ig

GRAPH = ig.Graph.Formula(",".join(DATA))

fig, ax = plt.subplots(figsize=(10,10))
ig.plot(
    GRAPH,
    target=ax,
    edge_width=0.1,
    vertex_size=0.2
)

plt.show()
```

## Identify Triads

```python
count = 0

for t in GRAPH.list_triangles():
    has_t = False
    for idx in t:
        if GRAPH.vs[idx]["name"].startswith("t"):
            has_t = True
            break
    
    if has_t:
        count += 1

print(f"count = {count}")
```
